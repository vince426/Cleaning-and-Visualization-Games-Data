### Popular Video Games 1980 - 2023

```python
import numpy as np
import pandas as pd
```

```python
games = pd.read_csv('games.csv',
                    parse_dates=['Release Date'])
```

```python
games.head()
```

```python
games.info()
```

```python
games.loc[games['Release Date'] == 'releases on TBD']
```

```python
# games.drop(games.loc[games['Release Date'] == 'releases on TBD'], axis='rows')
games = games[games['Release Date'] != 'releases on TBD']
```

```python
games.loc[games['Release Date'] == 'releases on TBD']
```

```python
# now we convert the date
games['Release Date'] = pd.to_datetime(games['Release Date'])
```

```python
games['Release Date'] = games['Release Date'].dt.strftime('%Y-%m-%d')
```

```python
games.head()
```

Now that the dates are set, we now have to clean the "Team" column

```python
# We have to remove the brackets in the ['Team'] column and also the apostrophes
games['Team'] = games['Team'].str.replace('[\[\]]', '', regex=True)
games['Team'] = games['Team'].str.replace("'", '', regex=True)
```

```python
games.head()
```

We also need to clean the ['Genres'] column using the same method we used in ['Team'] column

```python
# We have to remove the brackets in the ['Genres'] column and also the apostrophes
games['Genres'] = games['Genres'].str.replace('[\[\]]', '', regex=True)
games['Genres'] = games['Genres'].str.replace("'", '', regex=True)
```

```python
games.head()
```

we now have to split each genre in their own columns

```python
genres_split = games['Genres'].str.split(',', expand=True)
```

```python
genres_split.head()
```

It seems that a row went up to 7th genre, we have to deal with that later

```python
genres_split.columns = ['Genre 1', 'Genre 2', 'Genre 3', 'Genre 4', 'Genre 5', 'Genre 6', 'Genre 7']
```

```python
genres_split.head()
```

```python
# we have to drop genre 3 to genre 7 since it's a little bit too much
genres_split.drop(columns=['Genre 3', 'Genre 4', 'Genre 5', 'Genre 6', 'Genre 7'])
```

```python
games = pd.concat([games, genres_split], axis=1)
```

```python
games.head()
```

```python
# we now have to drop the column genres and reviews since it's unneccesary at this point
games = games.drop(columns=['Genres', 'Reviews'])
```

```python
games.head()
```

```python
games = games.drop(columns=['Genre 3', 'Genre 4', 'Genre 5', 'Genre 6', 'Genre 7'])
```

```python
games.head()
```

We now have to convert values with K into *1000 while retaining other values

```python
games['Times Listed Clean'] = games['Times Listed'].str.replace('K', '').astype(float)
games.loc[games['Times Listed'].str.contains('K'), 'Times Listed Clean'] *= 1000
games['Times Listed Clean'] = games['Times Listed Clean'].astype(int)
```

```python
games['Number of Reviews Clean'] = games['Number of Reviews'].str.replace('K', '').astype(float)
games.loc[games['Number of Reviews'].str.contains('K'), 'Number of Reviews Clean'] *= 1000
games['Number of Reviews Clean'] = games['Number of Reviews Clean'].astype(int)
```

```python
games['Plays Clean'] = games['Plays'].str.replace('K', '').astype(float)
games.loc[games['Plays'].str.contains('K'), 'Plays Clean'] *= 1000
games['Plays Clean'] = games['Plays Clean'].astype(int)
```

```python
games['Backlogs Clean'] = games['Backlogs'].str.replace('K', '').astype(float)
games.loc[games['Backlogs'].str.contains('K'), 'Backlogs Clean'] *= 1000
games['Backlogs Clean'] = games['Backlogs Clean'].astype(int)
```

```python
games['Wishlist Clean'] = games['Wishlist'].str.replace('K', '').astype(float)
games.loc[games['Wishlist'].str.contains('K'), 'Wishlist Clean'] *= 1000
games['Wishlist Clean'] = games['Wishlist Clean'].astype(int)
```

```python
games.head()
```

```python
# we now have to drop unclean columns
games = games.drop(columns=['Times Listed', 'Number of Reviews', 'Plays', 'Backlogs', 'Wishlist'])
```

```python
games
```

```python
# we have to drop 'playing' column since it's unneccesary
games = games.drop(columns='Playing')
```

```python
games.head()
```

```python
games.info()
```

```python
games.drop(columns='Unnamed: 0')
```
